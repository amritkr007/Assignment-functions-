1.What is the difference between a function and a method in Python?
- FUNCTION
  - A function is a standalone block of code that performs a specific task.

  - It is defined using the def keyword and can be called independently.

EXAMPLE


```python
def greet(name):
    return f"Hello, {name}!"

print(greet("Alice"))  # Output: Hello, Alice!
```

METHOD
  - A method is a function that is associated with an object and is called on that object.

  - Methods are defined inside classes and typically operate on the class instance (self).

EXAMPLE

```python
class Person:
    def __init__(self, name):
        self.name = name

    def greet(self):
        return f"Hello, {self.name}!"

p = Person("Alice")
print(p.greet())  # Output: Hello, Alice!
```

```python
def outer():
    def inner():
        return "Hello from inner!"
    return inner  # Returns the function itself

fn = outer()  # fn now holds the inner function
print(fn())   # Output: Hello from inner!
```

6. Explain the concept of generators in Python and how they are defined.
  - In Python, generators are a special type of iterable that allow you to iterate over data lazily, meaning they generate values one at a time as needed, instead of storing them all in memory at once. This makes them memory-efficient and useful for handling large datasets or infinite sequences.

DEFINING GENERATOR

Generators in Python can be defined in two main ways:

  - 1.Using Functions with yield
    - A generator function is a regular function that uses the yield keyword instead of return to return values one at a time.
    - When the function is called, it returns a generator object but does not execute the function immediately.
    - Execution starts when next() is called on the generator object, and it runs until it hits a yield statement.
    - The functionâ€™s state (variables, execution position) is preserved between calls.

```python
def count_up_to(n):
    count = 1
    while count <= n:
        yield count  # Produces a value and pauses execution
        count += 1

gen = count_up_to(5)  # Creates a generator object

print(next(gen))  # 1
print(next(gen))  # 2
print(next(gen))  # 3
```

2.Using Generator Expressions (Similar to List Comprehensions)
  - A generator expression is like a list comprehension but uses parentheses () instead of square brackets [].
  - It generates values lazily without storing the entire list in memory.

```python
gen_exp = (x**2 for x in range(5))

print(next(gen_exp))  # 0
print(next(gen_exp))  # 1
print(next(gen_exp))  # 4
```

KEY FEATURES FOR GENERATORS
- Lazy Evaluation: Values are generated on the fly, reducing memory usage.
- State Retention: Generators remember their last execution state, making them useful for tasks like streaming data processing.
- Iteration: Generators can be iterated using loops (for, while) or functions like next().
- Automatically Stop: When the generator is exhausted, it raises a StopIteration exception.

Generators are commonly used for tasks like:
  - Processing large files line by line.
  - Implementing infinite sequences.
  - Streaming data processing.

4.Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True
primes = [num for num in range(1, 201) if is_prime(num)]
print(primes)
```

5.Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
class FibonacciIterator:
    def __init__(self, n_terms):
        self.n_terms = n_terms
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.n_terms:
            raise StopIteration
        fib_number = self.a
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return fib_number

fib_iterator = FibonacciIterator(10)
for num in fib_iterator:
    print(num, end=" ")
```

6.Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_two(max_exponent):
    for exp in range(max_exponent + 1):
        yield 2 ** exp

for power in powers_of_two(5):
    print(power, end=" ")
```

7.Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_file_lines(filepath):

    try:
        with open(filepath, 'r') as file:
            for line in file:
                yield line.rstrip('\n')
    except FileNotFoundError:
        yield f"File not found: {filepath}"
    except Exception as e:
        yield f"An error occurred: {e}"


with open("my_file.txt", "w") as f:
    f.write("This is line 1.\n")
    f.write("This is line 2.\n")
    f.write("This is line 3.")

for line in read_file_lines("my_file.txt"):
    print(line)

for line in read_file_lines("non_existent_file.txt"):
    print(line)
```
